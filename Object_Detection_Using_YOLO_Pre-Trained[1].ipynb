```python
import cv2
from ultralytics import YOLO

# Load the YOLO model
model = YOLO("yolov8n.pt")

# Open the video capture (0 is the default camera)
cap = cv2.VideoCapture(0)

while True:
    # Capture frame-by-frame
    ret, frame = cap.read()
    if not ret:
        break

    # Make predictions on the current frame
    results = model.predict(frame)

    # Draw bounding boxes on the frame
    for result in results:
        for box in result.boxes:
            # Extract bounding box coordinates and convert to integers
            x1, y1, x2, y2 = map(int, box.xyxy[0].tolist())

            # Extract class and confidence, convert to float
            cls = box.cls.item() if isinstance(box.cls, torch.Tensor) else box.cls
            conf = box.conf.item() if isinstance(box.conf, torch.Tensor) else box.conf

            # Draw the bounding box on the frame
            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)

            # Add label and confidence score
            label = f"Class {int(cls)}: {conf:.2f}"
            cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    # Display the frame with predictions
    cv2.imshow('YOLOv8 Video Detection', frame)

    # Press 'q' to exit the loop
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# When everything is done, release the capture
cap.release()
cv2.destroyAllWindows()
```

```python
# Load the YOLO model
model = YOLO("yolov8n.pt")
classes=model.names
```

```python
# Open the video capture (0 is the default camera)
cap = cv2.VideoCapture(0)

while True:
    # Capture frame-by-frame
    ret, frame = cap.read()
    if not ret:
        break

    # Convert frame to RGB (YOLO might expect RGB format)
    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # Make predictions on the current frame
    results = model.predict(frame_rgb)

    # Draw bounding boxes on the frame
    for result in results:
        for box in result.boxes:
            # Extract bounding box coordinates and convert to integers
            x1, y1, x2, y2 = map(int, box.xyxy[0].tolist())

            # Extract class and confidence, convert to float
            cls = box.cls.item() if isinstance(box.cls, torch.Tensor) else box.cls
            conf = box.conf.item() if isinstance(box.conf, torch.Tensor) else box.conf

            # Draw the bounding box on the frame
            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)

            # Add label and confidence score
            label = f"Class {int(cls)}: {conf:.2f}"
            class_name = classes[int(cls)]
            cv2.putText(frame, label+" "+class_name, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    # Display the frame with predictions
    cv2.imshow('YOLOv8 Video Detection', frame)

    # Press 'q' to exit the loop
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# When everything is done, release the capture
cap.release()
cv2.destroyAllWindows()
```
